# ***üî•‚ú® EXPLORATORY DATA ANALYSIS (EDA) IN PYTHON
üîç What is EDA?
EDA stands for Exploratory Data Analysis, which is the first step in any data science or machine learning project.

üß† Why Do EDA?
Understand the structure of your data

Identify patterns, relationships, and anomalies

Clean the data for better model performance

Gain insights before modeling
The Breast Cancer Wisconsin dataset contains medical measurements of cell nuclei from breast mass biopsies.
The goal of EDA here is to understand the data structure, detect patterns, spot anomalies, and prepare the dataset for building a classification model that predicts whether a tumor is benign (B) or malignant (M).

üìÇ Dataset Overview
Samples (Rows): ~569

Features (Columns): ~30 numeric features + ID + diagnosis

Target Variable: diagnosis ‚Üí M (Malignant), B (Benign)

Each feature represents a measurement of a cell nucleus, such as:

Radius

Texture

Perimeter

Area

Smoothness

... (mean, worst, standard error)

***

# ***üóÉÔ∏è 1. Understanding the Dataset
Look at the structure: rows, columns, data types

Understand the features: what they represent

Identify target vs. predictors

üî† 2. Data Types
Numeric: Continuous or discrete numbers

Categorical: Labels or classes (e.g., diagnosis)

Date/Time (if applicable)

Knowing types helps decide which analysis or visualizations to use.

üßº 3. Missing Values
Missing values can cause errors or bias

Common ways to handle them:

Remove rows/columns

Fill with mean/median/mode

Impute based on other features

üìä 4. Univariate Analysis (One variable)
Helps understand distribution of each feature

Detects skewness, outliers, and spread

Visual tools: Histograms, Boxplots

üë´ 5. Bivariate Analysis (Two variables)
Studies relationship between two features

Useful for:

Checking correlation

Understanding how target varies with features

Visual tools: Scatter plots, Line plots

üî• 6. Correlation Analysis
Shows how features are related to each other

Helps detect:

Redundant features

Strong predictors

Typically shown using a heatmap

‚ö†Ô∏è 7. Outlier Detection
Outliers are values that are unusually high or low

May indicate:

Data entry errors

Rare but important cases

Noise in data

Visual tools: Boxplots, Histograms

üßΩ 8. Data Cleaning
Remove duplicates

Drop irrelevant columns (like IDs)

Fix inconsistencies (like wrong labels or formats)

Clean data is key for accurate analysis and modeling.

üìà 9. Trend Observation (Optional)
Use line plots or time-based charts to observe changes in data over samples or time.

Useful when data has order or time component.

# ****10.SCATTER CHART***

üåü What is a Scatter Plot?
A scatter plot is a type of graph that shows two features (variables) of your data using dots.

Each dot represents one row (sample) in your dataset.

üìå Example (Breast Cancer Dataset)
Let‚Äôs say you want to compare:

radius_mean = the average size of the tumor (x-axis)

texture_mean = the texture of the tumor (y-axis)

Each dot on the plot shows:

A single tumor

Its size on the x-axis

Its texture on the y-axis



```python

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='radius_mean', y='texture_mean', hue='diagnosis', palette='Set1')
plt.title("Scatter Plot of Radius vs Texture")
plt.xlabel("Radius Mean")
plt.ylabel("Texture Mean")
plt.show()
```

* What is a Line Plot?
A line plot connects data points with lines to show trends or changes in values.

It‚Äôs usually used when your x-axis is ordered, like time or index.

‚úÖ When to Use It?
To show trends over samples (index)

To compare how a value changes

To view smoothness or fluctuations in a numeric column

*

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x=range(len(df)), y=df['radius_mean'], label='Radius Mean')
sns.lineplot(x=range(len(df)), y=df['texture_mean'], label='Texture Mean')
plt.title("Line Plot: Radius vs Texture Mean")
plt.xlabel("Sample Index")
plt.ylabel("Value")
plt.legend()
plt.show()
```

# **10.HEAT MAP*

üî• What is a Heatmap?
A heatmap is a color-coded table that shows the correlation (relationship) between different numeric features.

The darker or brighter the color, the stronger the relationship between two features.

üìà In Breast Cancer Dataset
We use a heatmap to see:

Which features are strongly correlated (e.g., radius_mean and perimeter_mean)

Whether some features give duplicate info

Which features might be useful for classification

```python
plt.figure(figsize=(20,10))
c=df.corr(numeric_only=True)
sns.heatmap(c,cmap="BrBG",annot=True)
plt.show()
```
